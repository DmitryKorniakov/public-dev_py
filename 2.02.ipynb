```python
# 3. Задание: использование sys и os

import sys
import os
import time

def linux_analysis():
    """
    Упрощенный анализ для Linux
    """
    print("АНАЛИЗ LINUX")
    print("=" * 60)
    
    # 1. Файлы в текущей директории
    print("\n1. ФАЙЛЫ В ТЕКУЩЕЙ ДИРЕКТОРИИ")
    print("-" * 40)
    
    current_dir = os.getcwd()
    files_info = []
    
    for item in os.listdir(current_dir):
        item_path = os.path.join(current_dir, item)
        if os.path.isfile(item_path):
            try:
                stat = os.stat(item_path)
                files_info.append({
                    'name': item,
                    'size': stat.st_size,
                    'atime': stat.st_atime
                })
            except OSError:
                continue
    
    # Сортируем по времени доступа
    files_info.sort(key=lambda x: x['atime'], reverse=True)
    
    total_size = sum(f['size'] for f in files_info)
    print(f"Файлов: {len(files_info)}")
    print(f"Общий размер: {total_size / (1024*1024):.2f} MB")
    
    print("\nПоследние 10 файлов:")
    for file_info in files_info[:10]:
        atime_str = time.strftime('%d.%m.%Y %H:%M', time.localtime(file_info['atime']))
        size_kb = file_info['size'] / 1024
        print(f"  {file_info['name']:30} {size_kb:6.0f} KB  {atime_str}")
    
    # 2. Расширения файлов
    print("\n2. РАСШИРЕНИЯ ФАЙЛОВ")
    print("-" * 30)
    
    extensions = set()
    for file_info in files_info:
        _, ext = os.path.splitext(file_info['name'])
        if ext:
            extensions.add(ext.lower())
    
    if extensions:
        print("Найдены:", ", ".join(sorted(extensions)))
    else:
        print("Расширения не найдены")
    
    # 3. Память процесса (Linux)
    print("\n3. ПАМЯТЬ ПРОЦЕССА")
    print("-" * 30)
    
    pid = os.getpid()
    print(f"PID: {pid}")
    
    # Память из /proc для Linux
    try:
        with open(f"/proc/{pid}/status", 'r') as f:
            for line in f:
                if line.startswith('VmRSS:'):
                    rss_kb = int(line.split()[1])
                    print(f"Используемая память: {rss_kb / 1024:.1f} MB")
                    break
    except:
        print("Не удалось получить информацию о памяти")

if __name__ == "__main__":
    linux_analysis()
```
