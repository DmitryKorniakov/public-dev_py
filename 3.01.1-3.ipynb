```python
### Задание 1.3: Классификация треугольников

a = int(input('Введите значение первой стороны треугольника:\n'))

b = int(input('Введите значение второй стороны треугольника:\n'))

c = int(input('Введите значение третьей стороны треугольника:\n'))

sides = sorted([a, b, c])
if sides[0] + sides[1] < sides[2]:
    print('Не существует треугольник с такими сторонами')
else:
    if len(set(sides)) == 1:
        print('Равносторонний')
    elif len(set(sides)) == 2:
        print('Равнобедренный')
    else:
        print('Разносторонний')
```

```python
### Задание 2.2: Поиск простых чисел

limit = int(input('Введите значение, до какого числа будут выводиться простые числа:\n'))

def find_primes(n):
    primes = []
    for n1 in range(2, n + 1):
        # поочередно проверять делится ли число на число от 2 до самого числа
        for i in range(2, n1):
            if n1 % i == 0:
                break
        # если при прохождении цикла не выполнился брейк, то число записывается
        else:
            primes.append(n1)
    return primes

print(find_primes(limit))
```

```python
### Задание 2.4: Анализ текста

text = input('Введите текст на проверку\n')

symbol_count = len(text)
words = text.split()
words_count = len(words)

# Подсчет предложений (упрощенный вариант)
sentense_count = 0
for char in text:
    if char in '.!?':
        sentense_count += 1
if sentense_count == 0 and text.strip():
    sentense_count = 1

print(f'В тексте {symbol_count} символов, {words_count} слов и {sentense_count} предложений')

# Очистка слов
cleaned_words = []
for word in words:
    cleaned_word = word.strip('.,!?;:"\'()').lower()
    if cleaned_word:
        cleaned_words.append(cleaned_word)

word_freq = {}
for word in cleaned_words:
    word_freq[word] = word_freq.get(word, 0) + 1

sorted_words = sorted(word_freq.items(), key=lambda item: item[1], reverse=True)
print("Топ 5 слов:")
for word, count in sorted_words[:5]:
    print(f"{word}: {count}")
```

```python
### Задание 3.3: Кастомные исключения

class InsufficientFundsError(Exception):
    pass
class InvalidAccountError(Exception):
    pass
class TransactionLimitError(Exception):
    pass

def add_to_dict_immutable(d: dict, key, value):
    new_dict = d.copy()
    new_dict[key] = value
    return new_dict

class BankAccount:
    accounts = {}
    def __init__(self, account_number, balance=0):
        self.account_number: str = account_number  # тип строка для номера счета
        self.balance: float = balance  # тип float для баланса
        BankAccount.accounts[self.account_number] = self
        self.transaction_count: int = 0  # тип int для счетчика транзакций
        self.transaction_limit: int = 100  # тип int для лимита транзакций

    def deposit(self, amount):
        self.balance += amount
        if amount <= 0:
            raise ValueError('Сумма должна быть положительной')

    def withdraw(self, amount):
        if self.balance < amount:
            raise InsufficientFundsError('Недостаточно средств для снятия')
        if self.transaction_count >= self.transaction_limit:
            raise TransactionLimitError('Превышен лимит транзакций в месяц')
        self.transaction_count += 1
        if amount <= 0:
            raise ValueError('Сумма должна быть положительной')


        self.balance -= amount

    def transfer(self, amount, target_account):
        if self.balance < amount:
            raise InsufficientFundsError('Недостаточно средств для перевода')
        if target_account.account_number not in BankAccount.accounts:
            raise InvalidAccountError('Неверный номер счёта')
        if self.transaction_count >= self.transaction_limit:
            raise TransactionLimitError('Превышен лимит транзакций в месяц')
        self.balance -= amount
        target_account.balance += amount
        self.transaction_count += 1
        if amount <= 0:
            raise ValueError('Сумма должна быть положительной')
```
